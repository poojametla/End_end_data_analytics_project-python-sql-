```python
# read data from file and handle null values
import pandas as pd

df = pd.read_csv(r'C:\Users\sivak\orders_extracted\orders.csv', na_values=['Not Available', 'unknown'])

df.head(20)
df['Ship Mode'].unique()
```

```python
df['Ship Mode'].unique()
```

```python
# rename column names,make them lower case and replace with underscore
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
# derive new columns discount,sale price and profit
#df.columns

df['discount']= df['list_price']*df['discount_percent']*.01
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df.head(5)
```

```python
# convert order date from object data type to datetime
#df.dtypes
pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```
